```python
import pandas as pd
import numpy as np
import plotly.offline as py
import plotly.graph_objs as go
py.init_notebook_mode(connected=True)

df = pd.read_csv("bfi.csv")

print(df)

#remove nan element
dfE1 = df.dropna(subset=['E1']) #don't speak
#dfE4 = df.dropna(subset=['E4']) #make friends easily

E1=dfE1['E1'].values
print(E1)

E4=dfE1['E4'].values
print(E4)


trace = go.Scatter(x = E1, y = E4,  
                   mode = 'markers', 
                   marker = {'color': 'rgb(255,0,0)', 'symbol': 'circle', 'opacity': 0.5}) 

layout = go.Layout( title = 'Scatterplot (x vs y)',
                    xaxis = {'title': 'x'},
                    yaxis = {'title': 'y'})
fig = go.Figure(data = [trace], layout = layout)
py.iplot(fig)
```

```python
data = []
for val in [1.,2.,3.,4.,5.,6.]:
    num_g = E1[E4==val]
    trace = go.Box(x=num_g, name=val)
    data.append(trace)

layout = go.Layout(
    xaxis=dict(title="x"),
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

```python
def compare_histograms(a, b):
    bins = {"start":0, "end":+80, "size":0.5}
    data = [go.Histogram(x=np.array(a), xbins=bins, opacity=0.4, name="Don't talk a lot"),
            go.Histogram(x=np.array(b), xbins=bins, opacity=0.4, name='Make friends easily')]
    layout = go.Layout(title = 'Speak vs Friends', xaxis = {'title': 'Grade'}, yaxis = {'title': 'Number of people'}, barmode='overlay')
    fig = go.Figure(data, layout)
    py.iplot(fig)
    
compare_histograms(E1, E4)
```
